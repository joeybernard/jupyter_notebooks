```python
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib notebook
from datetime import datetime

data = pd.read_csv("data_test.csv")

x = data['x']
y = data['y']
z = data['z']
t = data['t']

Bx = data['Bx']
By = data['By']
Bz = data['Bz']
B = np.sqrt(Bx**2 + By**2 + Bz**2)
```

```python
fig1 = plt.figure()
ax1 = fig1.gca(projection='3d')
ax1.quiver(x, y, z, Bx.array, By.array, Bz.array, length=5)
#ax.scatter3D(x,y,z,c=B)

plt.show()
```

```python
fig2 = plt.figure()
ax2 = fig2.gca(projection='3d')
#grad = np.gradient(B, x.array, y.array, z.array)
#ax2.scatter3D(x, y, z, grad)
ax2.scatter3D(x,y,z,c=B)

plt.show()
```
